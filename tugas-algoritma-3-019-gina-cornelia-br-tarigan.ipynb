```python
harga = [5000,6000,7500,12000,8000]
i=0

while i < len (harga) :
    harga[i] = 'Rp' + str(harga[i])
    i += 1

print ('Harga dalam format mata uang:',harga)

# Output Harga dalam format mata uang: ['Rp5000', 'Rp6000', 'Rp7500', 'Rp12000', 'Rp8000' ]
```
